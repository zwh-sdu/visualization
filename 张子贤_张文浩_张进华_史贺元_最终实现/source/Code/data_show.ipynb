```python
# 导入相关库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None

data = pd.read_csv('vgsales.csv')
data
```

```python
new_data = data[data["Platform"] == 2007]
new_data
```

```python
data_new = data[["Platform","NA_Sales","EU_Sales","JP_Sales"]] 
data_new
```

```python
game_platform = data['Platform'].value_counts().to_frame()
game_platform = game_platform[game_platform['Platform'] >= 100]
# 绘制
plt.figure(figsize=(15, 5))
sns.barplot(x = game_platform.index , y = 'Platform' ,data = game_platform)
plt.xlabel('platform')
plt.ylabel('Number of games released')
plt.title("Number of games released on different platforms")
plt.show()
```

```python
global_sales = data[["Platform","Global_Sales"]]
global_sales = global_sales[global_sales["Global_Sales"] > 10]
global_sales
```

```python
game_platform = data['Publisher'].value_counts().to_frame()
game_platform = game_platform[game_platform['Publisher'] >= 500]
# 绘制
plt.figure(figsize=(15, 5))
sns.barplot(x = game_platform.index , y = 'Publisher' ,data = game_platform)
plt.xlabel('Publisher')
plt.ylabel('Number of games released')
plt.title("Number of games released on different Publisher")
plt.show()
```
