```python
import numpy as np
from sklearn.manifold import TSNE
from time import time
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.decomposition import PCA

from sklearn import datasets
from sklearn.manifold import TSNE
```

```python
from keras.datasets import mnist
(X_train,Y_train),(X_test,Y_test) = mnist.load_data()
print(X_train.shape)
```

```python
X_train=X_train.reshape(60000,-1)
Y_train=Y_train.reshape(60000,-1)
print(X_train.shape)
print(Y_train.shape)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
Y_train=list(map(int,Y_train))
```

```python
def plot_embedding(data, label, title):
    x_min, x_max = np.min(data, 0), np.max(data, 0)
    data = (data - x_min) / (x_max - x_min)

    fig = plt.figure()
    ax = plt.subplot(111)
    for i in range(data.shape[0]):
        plt.text(data[i, 0], data[i, 1], str(label[i]),color=plt.cm.Set1(label[i] / 10.),fontdict={'weight': 'bold', 'size': 5})
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    return fig
```

```python
print('Computing t-SNE embedding')
tsne1 = TSNE(n_components=2, n_jobs=-1, perplexity=30)
t0 = time()
result1 = tsne1.fit_transform(X_train)
fig = plot_embedding(result1, Y_train,'t-SNE perplexity=30 (time %.2fs)'% (time() - t0))
plt.show(fig)
```

```python
import seaborn as sns
x_min,x_max = np.min(result1,0),np.max(result1,0)
data = (result1-x_min)/(x_max-x_min)
fig = plt.figure(figsize=(10,10))
# 表示画布整个输出，不分割成小块区域，图形直接输出在整块画布上
ax = plt.subplot(111)#那么这个figure就是个2*2的矩阵图，也就是总共有4个图，1就代表了第一幅图
sns.scatterplot(result1[:,0], result1[:,1], hue=Y_train, legend='full',palette='Spectral')
plt.show()
```
